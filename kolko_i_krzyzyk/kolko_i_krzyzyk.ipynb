```python
plansza = ["-", "-", "-", "-", "-", "-", "-", "-", "-" ]

def wyswietl_plansze():
    print(plansza[0] + plansza[1] + plansza[2])
    print(plansza[3] + plansza[4] + plansza[5])
    print(plansza[6] + plansza[7] + plansza[8])
   

def kombinacje_wygranej():

    if plansza[0] + plansza[1] + plansza[2] == "xxx":
        return True
    elif plansza[3] + plansza[4] + plansza[5] == "xxx":
        return True
    elif plansza[6] + plansza[7] + plansza[8] == "xxx":
        return True
    elif plansza[0] + plansza[3] + plansza[6] == "xxx":
        return True
    elif plansza[1] + plansza[4] + plansza[7] == "xxx":
        return True
    elif plansza[2] + plansza[5] + plansza[8] == "xxx":
        return True
    elif plansza[0] + plansza[4] + plansza[8] == "xxx":
        return True
    elif plansza[2] + plansza[4] + plansza[6] == "xxx":
        return True
    elif plansza[0] + plansza[1] + plansza[2] == "ooo":
        return True
    elif plansza[3] + plansza[4] + plansza[5] == "ooo":
        return True
    elif plansza[6] + plansza[7] + plansza[8] == "ooo":
        return True
    elif plansza[0] + plansza[3] + plansza[6] == "ooo":
        return True
    elif plansza[1] + plansza[4] + plansza[7] == "ooo":
        return True
    elif plansza[2] + plansza[5] + plansza[8] == "ooo":
        return True
    elif plansza[0] + plansza[4] + plansza[8] == "ooo":
        return True
    elif plansza[2] + plansza[4] + plansza[6] == "ooo":
        return True
    else:
        return False
    

def gra_kolko_krzyzyk():
    print("KÓŁKO I KRZYŻYK!")
    
    while kombinacje_wygranej() is False:
        gdzie_x = int(input("Podaj gdzie postawic 'X' (zakres liczb od 1 do 9): "))
        
        if gdzie_x not in range(1,10):
            gdzie_x = int(input("Podaj gdzie postawic 'X' (zakres liczb od 1 do 9): "))
            continue
        elif plansza[gdzie_x - 1] == 'o':
            gdzie_x = int(input("MIEJSCE ZAJĘTE!!\nPodaj gdzie postawic 'X' (zakres liczb od 1 do 9): "))
            continue  
        elif plansza[gdzie_x - 1] == 'x':
            gdzie_x = int(input("MIEJSCE ZAJĘTE!!\nPodaj gdzie postawic 'X' (zakres liczb od 1 do 9): "))
            continue
        if plansza[gdzie_x - 1] == '-':
            plansza[gdzie_x - 1] = "x"
            wyswietl_plansze()
        if kombinacje_wygranej() is True:
            print("Wygrał 'X'. Koniec gry.")
            break
        if plansza.count("-") ==0:
            print("Plansza jest pełna. Koniec gry.")
            break
        

        print("\n")
        gdzie_o = int(input("Podaj gdzie postawic 'O' (zakres liczb od 1 do 9): "))
        
        if gdzie_o not in range(1,10):
            gdzie_o = int(input("Podaj gdzie postawic 'O' (zakres liczb od 1 do 9): "))
            continue
        if plansza[gdzie_o - 1] == 'x':
            gdzie_o = int(input("MIEJSCE ZAJĘTE!!\nPodaj gdzie postawic 'O' (zakres liczb od 1 do 9): "))
            continue
        if plansza[gdzie_o - 1] == 'o':
            gdzie_o = int(input("MIEJSCE ZAJĘTE!!\nPodaj gdzie postawic 'O' (zakres liczb od 1 do 9): "))
            continue
        if plansza[gdzie_o - 1] == '-':
            plansza[gdzie_o - 1] = "o"
            wyswietl_plansze()
        if kombinacje_wygranej() is True:
            print("Wygrało 'O'. Koniec gry.")
            break
        if plansza.count("-") ==0:
            print("Koniec gry. Plansza jest pełna")
            break
        
        print("\n")
      

gra_kolko_krzyzyk()



```
